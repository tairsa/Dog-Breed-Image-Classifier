```python
# # Dog Breed Image Classifier using fastai

# ## Problem Definition

# **Task:** Multi-class image classification.  
# Given an image of a dog, the model predicts which **breed** it belongs to.

# In my dataset I use several dog breeds, for example:

# - Beagle  
# - Doberman  
# - Golden Retriever  
# - Labrador Retriever  
# - Shih-Tzu  

# (The exact list of classes is taken from the folder names.)

# ### Motivation

# - Automatic tagging of dog images in photo collections or social media.
# - Could be used in a mobile app that helps users recognize dog breeds.
# - A realistic computer-vision task with multiple classes and visually similar categories.

# ### Expected Challenges

# - Some dog breeds look very similar (e.g. Golden vs Labrador Retriever).
# - Large variation inside each breed (different colors, sizes, ages).
# - Images may have different lighting, backgrounds, and camera angles.
# - Sometimes the full dog is not visible (only the head or part of the body).
```

```python
from fastai.vision.all import *
from pathlib import Path

# Path to dataset (relative to the notebook)
DATA_PATH = Path('data')
DATA_PATH.ls()
```

```python
# ## Dataset Creation and Preparation

# I collected a subset of dog images and organized them so that:

# - Each **folder name** corresponds to a dog breed (e.g. `beagle`, `Doberman`, `golden_retriever`, ...).
# - All images of that breed are stored inside the corresponding folder.
# - I let fastai create a training/validation split automatically using `valid_pct=0.2`.

# To document the final dataset structure, I count how many images there are in each breed (folder) using Python.
```

```python
import os
import shutil
import random

TARGET_DATA_DIR = DATA_PATH / 'dogs_subset'

# List of breed folders
BREED_FOLDERS = [d.name for d in DATA_PATH.iterdir() if d.is_dir()]

TRAIN_RATIO = 0.8
N_IMAGES_PER_CLASS = 150
N_TRAIN = int(N_IMAGES_PER_CLASS * TRAIN_RATIO)

# Step 1: Remove old folders and create a new clean structure
if TARGET_DATA_DIR.exists():
    shutil.rmtree(TARGET_DATA_DIR)

TARGET_DATA_DIR.mkdir(parents=True, exist_ok=True)
(TARGET_DATA_DIR / 'train').mkdir()
(TARGET_DATA_DIR / 'test').mkdir()

print(f"Creating train and test folders in: {TARGET_DATA_DIR}")

# Step 2: Copy and split images for each breed
for breed_name in BREED_FOLDERS:
    source_dir = DATA_PATH / breed_name
    
    # Ensure there are enough images for this breed
    all_files = list(source_dir.glob('*.jpg'))
    if len(all_files) < N_IMAGES_PER_CLASS:
        print(f"Warning: Only {len(all_files)} images found for breed {breed_name}. Using all available images.")
    
    # Shuffle and select the desired number of images
    random.shuffle(all_files)
    selected_files = all_files[:N_IMAGES_PER_CLASS]
    
    # Split into train and test sets
    train_files = selected_files[:N_TRAIN]
    test_files = selected_files[N_TRAIN:]

    # Create target folders for this breed
    (TARGET_DATA_DIR / 'train' / breed_name).mkdir()
    (TARGET_DATA_DIR / 'test' / breed_name).mkdir()

    # Copy training images
    for f in train_files:
        shutil.copy(f, TARGET_DATA_DIR / 'train' / breed_name / f.name)
        
    # Copy test images
    for f in test_files:
        shutil.copy(f, TARGET_DATA_DIR / 'test' / breed_name / f.name)

print("Data preparation complete. Files are organized in the dogs_subset folder.")
```

```python
# ## Data Loading in fastai

# I use `ImageDataLoaders.from_folder`:

# - Each subfolder inside `DATA_PATH` represents one dog breed (one class).
# - The label for each image is taken from the folder name.
# - `valid_pct=0.2` creates a random 80% / 20% train/validation split.
# - `item_tfms=Resize(224)` resizes all images to a fixed size.
# - `batch_tfms=aug_transforms()` applies standard data augmentation.
```

```python
NEW_DATA_PATH = Path('data/dogs_subset')
dls = ImageDataLoaders.from_folder(
    NEW_DATA_PATH,
    train='train',
    valid='test',
    item_tfms=Resize(224),
    batch_tfms=aug_transforms(),
    bs=64
)

dls.show_batch(max_n=9, figsize=(8, 8))
```

```python
# ## Model Training

# I use transfer learning with a pre-trained `ResNet34` model:

# - The convolutional backbone is pre-trained on ImageNet.
# - The final layers are replaced with a small head for 2-way classification.
# - I fine-tune the model for a few epochs using `learn.fine_tune`.
```

```python
learn = vision_learner(dls, resnet34, metrics=accuracy)

learn.fine_tune(3)  #     
```

```python
learn.export('dogs_resnet34.pkl')
```

```python
# Confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), cmap='YlOrBr')
```

```python
# Top losses: where the model was most wrong
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_top_losses(9, nrows=4, figsize=(10,10))
```

```python
learn.show_results(ds_idx=1, max_n=9, figsize=(8, 8))
```

```python
val_loss, val_acc = learn.validate()
val_loss, val_acc
```
